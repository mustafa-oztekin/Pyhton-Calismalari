```python
import psycopg2
import pandas as pd
df = pd.read_excel("rvmucret.xlsx")
```

```python
type(df)
```

```python
con = psycopg2.connect(
                        host = "localhost",
                        database = "my_database",
                        user = "postgres",
                        password = "1234")

cur = con.cursor()

for i in range(len(df)):
    cur.execute("insert into reward (kod, ucret) values (%s, %s)",
                (int(df.iloc[i,0]), df.iloc[i,1]))
    
# numpy.int64 old. icin int'e cevirdim.
# numpy.float64 old. icin int'e cevirdim.
# digerleri ya str ya da float old. icin tip degistimek gerekli degil.
# numpy.int64,float64'ten dogrudan %s yapamiyoruz.
cur.execute("select * from reward")
rows = cur.fetchall()
con.commit()
print(rows)
cur.close()
con.close()
```
